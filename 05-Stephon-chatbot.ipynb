# Demonstrate Seq2Seq Wrapper with twitter chat log

```python
import tensorflow as tf
import numpy as np

# preprocessed data
import sys
sys.path.append("./datasets/twitter/")
import data
import data_utils
```

```python
# load data from pickle and npy files
metadata, idx_q, idx_a = data.load_data(PATH='datasets/stephone/')
(trainX, trainY), (testX, testY), (validX, validY) = data_utils.split_dataset(idx_q, idx_a)
```

```python
# parameters 
xseq_len = trainX.shape[-1]
yseq_len = trainY.shape[-1]
batch_size = 16
xvocab_size = len(metadata['idx2w'])  
yvocab_size = xvocab_size
emb_dim = 1024
```

```python
import seq2seq_wrapper
```

```python
model = seq2seq_wrapper.Seq2Seq(xseq_len=xseq_len,
                               yseq_len=yseq_len,
                               xvocab_size=xvocab_size,
                               yvocab_size=yvocab_size,
                               ckpt_path='ckpt/twitter/',
                               emb_dim=emb_dim,
                               num_layers=5
                               )
```

```python
val_batch_gen = data_utils.rand_batch_gen(validX, validY, 256)
test_batch_gen = data_utils.rand_batch_gen(testX, testY, 256)
train_batch_gen = data_utils.rand_batch_gen(trainX, trainY, batch_size)
```

```python
sess = model.train(train_batch_gen, val_batch_gen,sess=sess)
```

```python
import jieba
def talk( l = '今天有想要去哪裡嗎qmark'):
    
    l = l.replace('！','EXMARK')
    l = l.replace('!','EXMARK')
    l = l.replace('？','QMARK')
    l = l.replace('?','QMARK')
    l = l.lower()
    
    testq = l
    test_array = []

    for w in jieba.cut(testq):
        w = w.encode('utf-8')
        if w in metadata['w2idx']:
#             print metadata['w2idx'][w]
            test_array.append( metadata['w2idx'][w])
        else:
#             print metadata['w2idx']['UNK']
            test_array.append( metadata['w2idx']['unk'])

    test_array = test_array +  [0]*20
    # print test_array
    test_array = test_array[:20]
    test_array = np.array([test_array])
    result = model.predict(sess, test_array.T)
    
    replies = []

    for ii, oi in zip(test_array, result):


        qs = ''
        for w in data_utils.decode(sequence=ii, lookup=metadata['idx2w'], separator=' ').split(' '):
            qs += w

        qs = qs.replace("dot",' ')
        qs = qs.replace("qmark",'?')
        qs = qs.replace("exmark",'!')
        print 'question:', qs

        anss = ''
        for w in data_utils.decode(sequence=oi, lookup=metadata['idx2w'], separator=' ').split(' '):
            anss+= w

        anss = anss.replace("dot",' ')
        anss = anss.replace("qmark",'?')
        anss = anss.replace("exmark",'!')

        print 'answer:', anss
        print 
        return anss
```
